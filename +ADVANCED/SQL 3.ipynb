### Store the dataset in a Table

##### In many cases the dataset to be analyzed is available as a .CSV (comma separated values) file, perhaps on the internet. To analyze the data 
##### using SQL, it first needs to be stored in the database.

##### We will first read the csv files  from the given url  into pandas dataframes

##### Next we will be using the  df.to_sql() function to convert each csv file  to a table in sqlite  with the csv data loaded in it.


```python
import pandas
df = pandas.read_csv('ChicagoPublicSchools.csv')
df.to_sql("ChicagoPublicSchools", con, if_exists='replace', index=False, method="multi", chunksize=300)
```

### Problem 4

##### What are the top 10 schools with the highest "Average Student Attendance"?


```python
%%sql
SELECT AVERAGE_STUDENT_ATTENDANCE, NAME_OF_SCHOOL FROM ChicagoPublicSchools
GROUP BY AVERAGE_STUDENT_ATTENDANCE, NAME_OF_SCHOOL
ORDER BY AVERAGE_STUDENT_ATTENDANCE DESC LIMIT 10
```

### Problem 5

##### Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance


```python
%%sql
SELECT AVERAGE_STUDENT_ATTENDANCE, NAME_OF_SCHOOL FROM ChicagoPublicSchools
GROUP BY AVERAGE_STUDENT_ATTENDANCE, NAME_OF_SCHOOL
ORDER BY AVERAGE_STUDENT_ATTENDANCE ASC NULLS LAST LIMIT 5
```

### Problem 6

##### Now remove the '%' sign from the above result set for Average Student Attendance column


```python
%%sql
SELECT REPLACE(AVERAGE_STUDENT_ATTENDANCE, '%', ''), NAME_OF_SCHOOL FROM ChicagoPublicSchools
GROUP BY AVERAGE_STUDENT_ATTENDANCE, NAME_OF_SCHOOL
ORDER BY AVERAGE_STUDENT_ATTENDANCE ASC NULLS LAST LIMIT 5

```

### Problem 7

##### Which Schools have Average Student Attendance lower than 70%?


```python
%%sql
SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE FROM ChicagoPublicSchools
WHERE CAST(REPLACE(Average_Student_Attendance, '%', '') AS DOUBLE ) < 70
ORDER BY AVERAGE_STUDENT_ATTENDANCE
```

### Problem 8

##### Get the total College Enrollment for each Community Area


```python
%%sql
SELECT COMMUNITY_AREA_NAME, COUNT(COMMUNITY_AREA_NAME) FROM ChicagoPublicSchools
GROUP BY COMMUNITY_AREA_NAME
ORDER BY COUNT(COMMUNITY_AREA_NAME) DESC
```

### Problem 9

##### Get the 5 Community Areas with the least total College Enrollment  sorted in ascending order


```python
%%sql
SELECT COMMUNITY_AREA_NAME, COUNT(COMMUNITY_AREA_NAME) FROM ChicagoPublicSchools
GROUP BY COMMUNITY_AREA_NAME
ORDER BY COUNT(COMMUNITY_AREA_NAME) ASC LIMIT 5
```

### Problem 10

##### List 5 schools with lowest safety score.


```python
%%sql
SELECT NAME_OF_SCHOOL, SAFETY_SCORE FROM ChicagoPublicSchools
ORDER BY SAFETY_SCORE ASC NULLS LAST LIMIT 5
```

### Problem 11

##### Get the hardship index for the community area which has College Enrollment of 4368


```python
df = pandas.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')
df.to_sql("chicago_socioeconomic_data", con, if_exists='replace', index=False,method="multi")
```

```python
%%sql
SELECT * FROM chicago_socioeconomic_data LIMIT 5
```

```python
%%sql
SELECT C_P.community_area_name, C_S.hardship_index
FROM ChicagoPublicSchools AS C_P
JOIN chicago_socioeconomic_data AS C_S
ON LOWER(C_S.community_area_name) = LOWER(C_P.COMMUNITY_AREA_NAME)
WHERE C_P.COLLEGE_ENROLLMENT = 4368
GROUP BY C_P.community_area_name, C_S.hardship_index

```

### Problem 12

##### Get the hardship index for the community area which has the highest value for College Enrollment


```python
%%sql
SELECT C_P.community_area_name, C_S.hardship_index, C_P.COLLEGE_ENROLLMENT
FROM ChicagoPublicSchools AS C_P
JOIN chicago_socioeconomic_data AS C_S
ON LOWER(C_S.community_area_name) = LOWER(C_P.COMMUNITY_AREA_NAME)
GROUP BY C_P.community_area_name, C_S.hardship_index, C_P.COLLEGE_ENROLLMENT
ORDER BY C_P.COLLEGE_ENROLLMENT DESC LIMIT 1
```
